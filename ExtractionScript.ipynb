```python
#pip install --upgrade google-api-python-client google-auth-httplib2 google-auth-oauthlib
#pip install beautifulsoup4

from __future__ import print_function
import os.path
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError


SCOPES = ['https://www.googleapis.com/auth/gmail.readonly']


def main():

    
    #This code is for authentication and authorization
    creds = None
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run in a token file
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    try:
        # Connecting to the GmailAPI
        service = build('gmail', 'v1', credentials=creds)
        #Getting a list of 200 emails
        results = service.users().messages().list(maxResults = 200, userId='me').execute()
        messages = results.get('messages')

        for msg in messages: 
            # Get the message from its id 
            txt = service.users().messages().get(userId='me', id=msg['id']).execute()
            
            try: 
            # Get value of 'payload' from dictionary 'txt' 
                payload = txt['payload'] 
                headers = payload['headers'] 
  
            # Look for Subject and Sender Email in the headers 
                for d in headers: 
                    if d['name'] == 'Subject': 
                        subject = d['value'] 
                    if d['name'] == 'From': 
                        sender = d['value'] 
                    
                print("Sender : ", sender)
                print("Subject : ", subject)
                print('\n')
                
            except:
                print("There was an error in reading the emails")
                    
    except HttpError as error:
        # Handle errors from gmail API. In case the API Call fails
        print(f'An error occurred: {error}')


if __name__ == '__main__':
    main()
```
